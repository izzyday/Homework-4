```python
import numpy as np
import matplotlib.pyplot as plt

marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
x = ["andy", "amy", "james", "jules", "arthur"]
y = [88, 66, 90, 55, 77]
plt.bar(x, y, width = 0.5)
plt.title("Marks Bar Chart")
```

```python
import numpy as np
import matplotlib.pyplot as plt

marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
y = np.array([88, 66, 90, 55, 77])
mylabels = ["andy", "amy", "james", "jules", "arthur"]

plt.pie(y, labels = mylabels)
plt.title("Marks Pie")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc = 'upper left')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 5, 7, 9])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([5, 7, 9, 11])

plt.subplot(2, 3, 1)
plt.plot(x1, y1, x2, y2)
plt.title("incomes")

x = ["apples", "bananas", "oranges"]
y = [6, 4, 9]

plt.subplot(2, 3, 2)
plt.bar(x, y, width = 0.5)
plt.title("fruits eaten")

y = np.array([6, 8, 12, 13])
mylabels = ["math", "english", "history", "science"]
plt.subplot(2, 3, 3)
plt.pie(y, labels = mylabels)
plt.title("subjects")

x = np.array([1, 2, 3, 4])
y = np.array([100, 200, 300, 400])

plt.subplot(2, 3, 4)
plt.plot(x, y)
plt.grid()
plt.title("time spent studying")

x = np. array([1, 2, 3, 4, 5, 6])
y = np.array([88, 76, 50, 99, 77, 45])
plt.subplot(2, 3, 5)
plt.scatter(x, y)
plt.title("boys' height")


x = np.random.normal(175, 17, 300)
plt.subplot(2, 3, 6)
plt.hist(x)
plt.title("income over time")
plt.show()
```
